```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/BankNoteAuthentication.csv")
```

```python
data.head()
```

```python
data.drop(['class'],axis=1)
```

```python
x=data[['variance','skewness','curtosis','entropy']]
y=data['class']
```

```python
x.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

Decission Tree

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
```

```python
tree.fit(x_train,y_train)
```

```python
tree_predictions=tree.predict(x_test)
```

```python
# accuaracy of decisiion tree model
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,tree_predictions)
ac=accuracy_score(y_test,tree_predictions)
```

```python
print(cm)
```

```python
print(ac)
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
random_forest=RandomForestClassifier()
```

```python
random_forest.fit(x_train,y_train)
```

```python
forest_predictions=random_forest.predict(x_test)
```

```python
#accuracy of random forest model
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,forest_predictions)
ac=accuracy_score(y_test,forest_predictions)
```

```python
print(cm)
```

```python
print(ac)
```
